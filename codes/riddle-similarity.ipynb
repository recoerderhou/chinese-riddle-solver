直接暴力计算相似度

```python
from sentence_transformers import SentenceTransformer
import numpy as np
import pdb

#csvReader = csv.reader(open(path, encoding='ISO-8859-1'))
csvReader = csv.reader(open(path, encoding='gb18030'))

def cosine_similarity(x,y):
    num = x.dot(y.T)
    denom = np.linalg.norm(x) * np.linalg.norm(y)
    return num / denom

model = SentenceTransformer('sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2')

flag = 0
corr = [0 for i in range(14)]
cnt = [0 for i in range(14)]
for line in csvReader:
    if flag == 0:
        flag = 1
        continue
    
    #print(line[0])
    length = len(line[0])
    if length > 12:
        length = 13
        
    sentences = [line[i] for i in range(6)]
    embeddings = model.encode(sentences)
    
    score = []
    for i in range(1, 6):
        sim = cosine_similarity(embeddings[0], embeddings[i])
        score.append(sim)
        
    predict = score.index(max(score))
    if predict == int(line[6]):
        corr[length] += 1
        corr[0] += 1
    
    cnt[length] += 1 
    cnt[0] += 1
    
print("total .......... Acc =", corr[0]/cnt[0])
for i in range(1, 13):
    if cnt[i] != 0:
        print("len="+str(i)+" ........ cnt =", cnt[i], "........ Acc =", corr[i]/cnt[i])
print("len>12 ........ cnt =", cnt[13], "........ Acc =", corr[13]/cnt[13])
```
